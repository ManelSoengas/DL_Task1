**BERT**, siglas de "Bidirectional Encoder Representations from Transformers"

Este algoritmo diseñado por Google persigue los siguientes objetivos :


1.   Reconozca los términos clave de una búsqueda
2.   Entender el contexto


El algoritmo ha sido entrenado mediante las búsquedas de los usuarios han hecho en Google y documentos índice de Google.El tipo de dato utilizado para el entrenamiento depende del objetivo de la tarea del modelo. Es decir existen modelos prenentrenados de BERT para tareas específicas, análisis de sentimientos en twitter o de opiniones.

El algoritmo pondera la importancia relativa de las diferentes palabras en una oración. En lugar de procesar el texto en orden secuencial, BERT evalúa simultáneamente el contexto de cada palabra en relación con todas las demás palabras de la oración.




A continuación se pueden ver dos ejemplos de predicción del modelo BERT para dos idiomas, Inglés y Español.
Facilitamos una frase a la que le falta una palabra. El modelo se encarga de predecir que palabra es la más adecuada según el contexto. Puedes cambiar la frase para probar diferentes situaciones.

```python
from transformers import BertTokenizer, BertForMaskedLM
import torch

# Cargar el tokenizador y el modelo preentrenado
tokenizer = BertTokenizer.from_pretrained('bert-base-uncased')
model = BertForMaskedLM.from_pretrained('bert-base-uncased')

# Texto de ejemplo con una palabra enmascarada
text = "The capital of France is [MASK]."

# Tokenizar el texto
input = tokenizer.encode(text, return_tensors="pt")

# Predecir todas las tokens
with torch.no_grad():
    outputs = model(input)
    predictions = outputs[0]

# Encontrar la palabra enmascarada
masked_index = torch.where(input == tokenizer.mask_token_id)[1]

predicted_token_id = predictions[0, masked_index].argmax(axis=1)
predicted_token = tokenizer.decode(predicted_token_id)

print("Texto original:", text)
print("Predicción:", predicted_token)
```

```python
from transformers import BertTokenizer, BertForMaskedLM
import torch

# Cargar el tokenizador y el modelo preentrenado multilingüe
tokenizer = BertTokenizer.from_pretrained('bert-base-multilingual-cased')
model = BertForMaskedLM.from_pretrained('bert-base-multilingual-cased')

# Texto de ejemplo con una palabra enmascarada
text = "La capital de Francia es [MASK]."

# Tokenizar el texto
input = tokenizer.encode(text, return_tensors="pt")

# Predecir todas las tokens
with torch.no_grad():
    outputs = model(input)
    predictions = outputs[0]

# Encontrar la palabra enmascarada
masked_index = torch.where(input == tokenizer.mask_token_id)[1]

predicted_token_id = predictions[0, masked_index].argmax(axis=1)
predicted_token = tokenizer.decode(predicted_token_id)

print("Texto original:", text)
print("Predicción:", predicted_token)
```

# **¿Qué relación existe entre Chatgpt (modleo del lenguaje) y un modelo basado en transformers?**

Cuando ChatGPT recibe una sentencia o pregunta del usuario, utiliza su arquitectura de Transformer para analizar y entender el texto. Se enfoca tanto en el contenido explícito como en el contexto implícito de la entrada.

**Predicción de Texto**: Utilizando su entrenamiento previo y la arquitectura Transformer, ChatGPT genera una respuesta prediciendo la secuencia de palabras que tendría más sentido en el contexto dado. Esto se hace generando una palabra a la vez, cada una basada en las palabras anteriores y el contexto general de la conversación.

Además un aspecto clave es la capacidad del modelo para mantener la coherencia y el contexto a lo largo de una conversación. Esto significa que las respuestas no solo dependen de la última pregunta o sentencia, sino también de la conversación previa.
